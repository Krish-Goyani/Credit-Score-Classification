# Data Cleaning

```python
# cramersV for finding relation between to features
import pandas as pd
from scipy.stats import chi2_contingency

def cramersV(col) :
    contingency_table = pd.crosstab(df[col],df['Credit_Score'])
    chi2, p, dof, expected = chi2_contingency(contingency_table)

    num_rows, num_cols = contingency_table.shape
    n = contingency_table.sum().sum()
    V = (chi2 / (n * (min(num_rows, num_cols) - 1))) ** 0.5
    print( f" {col}s Cramér's V:  {V}")
```

```python
cramersV('Month')
```

```python
# first of all i don't required name,age,ssn,occupation,month column also it contians 9k null values,also age i find there is any relation between age and credit score,SSN
df.drop(axis=1,labels=['Name','Age','SSN','Month','Occupation'],inplace=True)
df.shape
```

```python
df.info()
```

```python
# by surfing dataset i found that there is many starnge values so first of all let's replace them all
def strange_values(col):
    df[col]=df[col].str.strip('_')
    df[col]=df[col].replace(['!@9#%8','', 'nan'],np.NaN)
```

```python
cols=list(df.select_dtypes(include='O').columns)
list(map(strange_values,cols))
```

```python
df.describe(exclude=np.number).T
```

Now it's look good

```python
#many features have wrong datatypes so let's make them right
#below function chnage the datatypes
def change_dtype(col,dtype):
    df[col]=df[col].astype(dtype)
```

```python
required_features=['Annual_Income','Num_of_Loan','Delay_from_due_date','Num_of_Delayed_Payment','Changed_Credit_Limit','Outstanding_Debt','Amount_invested_monthly','Monthly_Balance']
for col in required_features:
    change_dtype(col,'float64')
```

```python
# also converting credit history age to only month form
def month_converter(x):
    if pd.notnull(x):
        num1 = int(x.split(' ')[0])
        num2 = int(x.split(' ')[3])
      
        return (num1*12)+num2
    else:
        return x
```

```python
df['Credit_History_Age'] = df.Credit_History_Age.apply(lambda x: month_converter(x)).astype('float64')
```

# Let's Fight with null values

```python
df.info()
```

```python
df.groupby('Customer_ID')['Credit_Mix'].apply(list)[1]
```

We can see that by grouping dataset by customer id we can replave null values with most frequent values

```python
temp=df.copy()
```

```python
#below function group by the dataset by customer id and replave the null values with most frequent value in the list
def null_replace_bymf(col):
    grouped=df.groupby('Customer_ID')[col]
    df[col]=grouped.transform(lambda x: x.fillna(stats.mode(x)[0][0]))
    
    print("befor number of null values : ",temp[col].isnull().sum())
    print("after number of null values : ",df[col].isnull().sum())
```

```python
null_replace_bymf('Credit_Mix')
```

```python
null_replace_bymf('Payment_Behaviour')
```

```python
df['Type_of_Loan'].replace([np.NaN], 'No Data Available', inplace=True)
```

Here i have found that many data contains minus value that is not possible so replacing minus values with null

```python
# Function to replace negative values with NaN
def replace_negatives(value):
    if isinstance(value, (int, float)):
        return value if value >= 0 else np.NaN
    else:
        return value
```

```python
df = df.applymap(replace_negatives)
```

```python
temp=df.copy()
```

```python
#below function group by the dataset by customer id and replace the outliers or unexpected values and null values with most frequent value in the list
def mode_replace_outliers_values(col):
    grouped=df.groupby('Customer_ID')[col]
    df[col]=grouped.transform(lambda x: stats.mode(x,nan_policy='omit')[0][0])
    print(f"for {col}")
    print(f"befor max value : {temp[col].max()}")
    print(f"befor min value : {temp[col].min()}")
    print(f"before null values : {temp[col].isnull().sum()}")
    print(f"after max value : {df[col].max()} ")
    print(f"after min value : {df[col].min()}")
    print(f"after null values : {df[col].isnull().sum()}")

    print()
    print()   
```

```python
for col in num_features:
    mode_replace_outliers_values(col)
```

Now we have by grouping tha data by customerid and replacing all values with mode value removed unexcpected values also replaced null values with most mode values and now we have zero null values in our whole dataset.
